# Customer Churn Prediction - EDA y Modelado

Notebook guia para explorar el dataset Telco Customer Churn y conectar el analisis con el pipeline reproducible del proyecto.

````python
import sys
from pathlib import Path

ROOT = Path.cwd().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(ROOT / 'src'))

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from data_loader import load_data, clean_data

sns.set_theme(style='whitegrid', palette='Set2')
````

## 1. Carga de Datos

````python
df = clean_data(load_data())
df.head()
````

````python
df.info()
df.describe(include='all').T.head(15)
````

## 2. Balance de la Variable Objetivo

````python
display(df['Churn'].value_counts(normalize=True).rename('proportion'))
sns.countplot(data=df, x='Churn')
plt.title('Distribucion de Churn')
plt.show()
````

## 3. Variables Numericas

````python
numeric_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
df[numeric_cols].describe()
````

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, col in zip(axes, numeric_cols):
    sns.histplot(data=df, x=col, hue='Churn', kde=True, ax=ax)
    ax.set_title(f'{col} por Churn')
plt.tight_layout()
plt.show()
````

## 4. Variables Categoricas Clave

````python
contract_rate = (
    df.assign(churn_flag=df['Churn'].map({'No': 0, 'Yes': 1}))
      .groupby('Contract', as_index=False)['churn_flag']
      .mean()
      .sort_values('churn_flag', ascending=False)
)
sns.barplot(data=contract_rate, x='Contract', y='churn_flag')
plt.title('Tasa de churn por contrato')
plt.ylabel('Churn rate')
plt.show()
````
